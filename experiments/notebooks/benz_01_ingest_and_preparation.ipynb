# Translate language to English

```python
from langdetect import detect
from deep_translator import GoogleTranslator

def translate_to_english(text):
    try:
        lang = detect(text)

        if lang == "en":
            return text

        translated = GoogleTranslator(source=lang, target="en").translate(text)
        return translated

    except:
        return text  # fallback if detection fails
```

```python
jobs_df["description_en"] = jobs_df["description"].apply(translate_to_english)
```

# Cleaning Description

```python
jobs_list = jobs_df.description_en.to_list()
```

```python
import re

results = []

for idx, job in enumerate(jobs_list):

    text = job

    # -----------------------------
    # 1. TASK DETECTION
    # -----------------------------
    task_match = re.search(r'(?i)Tasks', text)

    # -----------------------------
    # 2. QUALIFICATION DETECTION (ALL YOUR PATTERNS IN ORDER)
    # -----------------------------
    qual_match = None

    patterns = [
        r'(?i)(?:^|[\s\n]|(?<=[a-z]))Qualifications(?:[\s\n]|(?=[A-Z])|$)',
        r'(?i)(?:^|[^A-Za-z0-9])Qualifications(?:$|[^A-Za-z0-9])',
        r'(?i)\bQualifications(?:Qualification)?\b',
        r'(?i)Qualifications(?:Qualification)?'   # fallback for merged OCR cases
    ]

    for p in patterns:
        qual_match = re.search(p, text)
        if qual_match:
            break

    # -----------------------------
    # DEFAULTS
    # -----------------------------
    task_text = text
    qual_text = None
    qual_flag = "Not found"

    # -----------------------------
    # CASE A: BOTH FOUND
    # -----------------------------
    if task_match and qual_match:

        task_text = text[task_match.end():qual_match.start()]
        qual_text = text[qual_match.end():]

        qual_flag = "Found"

    # -----------------------------
    # CASE B: ONLY TASK FOUND
    # -----------------------------
    elif task_match and not qual_match:

        task_text = text[task_match.end():]
        qual_text = None
        qual_flag = "Not found"

    # -----------------------------
    # CASE C: ONLY QUALIFICATION FOUND
    # -----------------------------
    elif qual_match and not task_match:

        task_text = text[:qual_match.start()]
        qual_text = text[qual_match.end():]
        qual_flag = "Found"

    # -----------------------------
    # CASE D: NONE FOUND
    # -----------------------------
    else:

        task_text = text
        qual_text = None
        qual_flag = "Not found"

    # -----------------------------
    # STORE RESULT
    # -----------------------------
    results.append({
        "index": idx,
        "tasks": task_text.strip(),
        "qualifications": qual_text.strip() if qual_text else None,
        "qualification_flag": qual_flag
    })

    print(idx, qual_flag)
```
