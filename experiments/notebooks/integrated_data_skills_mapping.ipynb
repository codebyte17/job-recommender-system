# Skills mapping

```python
# temporary code for help 
import ast

import ast
import numpy as np
import re

def safe_parse(x):
    try:
        return ast.literal_eval(x)
    except:
        # fix numpy types like np.int64(2)
        x = re.sub(r'np\.int64\((\d+)\)', r'\1', x)
        x = re.sub(r'np\.float64\(([\d.]+)\)', r'\1', x)
        return ast.literal_eval(x)

df['results'] = df['skills_ids'].apply(safe_parse)
```

```python
df.head()
```

```python
def map_skill_names(annotations):
    try : 
        #annotations = skill_extractor.annotate(description)
        skills_maped = set()
        for skill in annotations.get("full_matches", []):
            #skills.add(skill["doc_node_value"])
            skills_maped.add(SKILL_DB[skill["skill_id"]]["skill_name"])
    
        for skill in annotations.get("ngram_scored", []):
            if skill["score"] > 0.5: 
                #skills.add(skill["doc_node_value"])
                skills_maped.add(SKILL_DB[skill["skill_id"]]["skill_name"])
            else: 
                pass 
    
        return skills_maped
    except Exception as e: 
        return None 

df["skills_maped"] = df["results"].apply(map_skill_names)
```

```python
df.skills_maped.to_list()[1]
```

```python
df.drop(["skills_ids","results"],axis=1,inplace=True)
```

```python
df.head()
```

```python
def clean_employment_type(text):
    text = text.lower()

    # Replace separators with spaces
    text = text.replace('-', ' ')
    return text

df["job_type"] = df.job_type.apply(clean_employment_type)
```

```python
df['strata'] = (
    df['Department'].astype(str) + "_" +
    df['job_type'].astype(str) + "_" +
    df['location'].astype(str) + "_" +
    df['field_of_work'].astype(str)
)
```

```python
df_sample = df.groupby('strata', group_keys=False).apply(
    lambda x: x.sample(frac=0.35, random_state=42)
)
```
