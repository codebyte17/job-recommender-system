```python
import spacy

# Load spaCy ONCE (keep only sentence splitting)
nlp = spacy.load("en_core_web_sm", disable=["ner", "tagger", "lemmatizer"])
```

```python


def process_jobs(texts, labels, batch_size=32):
    """
    Fast batch NLP pipeline:
    1. Sentence split using spaCy (batched)
    2. Flatten sentences for transformer
    3. Run classifier in one batch
    4. Map results back to original texts
    """

    all_sentences = []
    mapping = []

    # STEP 1: sentence splitting (fast, no heavy components)
    for i, text in enumerate(texts):
        if not text:
            continue

        doc = nlp(text.replace("\n", " "))

        for sent in doc.sents:
            s = sent.text.strip()
            if len(s.split()) > 4:
                all_sentences.append(s)
                mapping.append(i)
    print("Step 1 completed! ")
    # If nothing found
    if not all_sentences:
        return [""] * len(texts)

    # STEP 2: run transformer classifier ONCE (BIG SPEED BOOST)
    outputs = classifier(
        all_sentences,
        candidate_labels=labels,
        batch_size=batch_size
    )

    # STEP 3: aggregate results per original text
    grouped = [[] for _ in range(len(texts))]

    for out, idx in zip(outputs, mapping):
        grouped[idx].append(out)

    print("Step 3 completed!")
    # STEP 4: your custom filtering function
    def filter_response(results):
        composed_sent = []
        for result in results:
            indexs = [result['labels'].index(field) for field in ["EXPERIENCE","REQUIREMENTS","QUALIFICATIONS","TOOLS_TECHNOLOGIES","SKILLS","EDUCATION","RESPONSIBILITY","TASK_TO_DO"]]
            scores = [result["scores"][idx] for idx in indexs]
            if sum(scores) > 0.55 : 
                #print(f"Relevant materials of category {result['labels'][result['scores'].index(max(result['scores']))]} is {max(result['scores'])} {sent}")
                composed_sent.append(str(result["sequence"]))
    
        return " ".join(composed_sent)

    return [filter_response(g) for g in grouped]
```
