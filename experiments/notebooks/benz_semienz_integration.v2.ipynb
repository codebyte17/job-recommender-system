# Rename and Extract important columns

```python
mbenz_df.rename(columns={
    "job_category" : "field_of_work",
    "working_mode" : "job_type", 
    "organization" : "company", 
    "publication"  : "posted_since"
},inplace=True)

semienz_df.rename(columns={
    "organization" : "Department",
    "clean_description" : "level1_clean",
    "final_job_description" : "clean_description",
    "job_title" : "title"
},inplace=True)
```

```python
mbz_df = mbenz_df[["job_id","title","field_of_work","posted_since","company","Department","job_type","location","clean_description"]].copy()
smnz_df = semienz_df[["job_id","title","field_of_work","posted_since","company","Department","job_type","location","clean_description"]].copy()
```

## Change the type of posted_since column to Datetime format

```python
mbz_df['posted_since'] = pd.to_datetime(mbz_df['posted_since'],format='mixed')
smnz_df['posted_since'] = pd.to_datetime(smnz_df['posted_since'], format='mixed')
```

## Combined benz and Semiens Job applications data

```python
integrated_df = pd.concat([mbz_df,smnz_df])
```

# Translate the Language at title level 

```python
from langdetect import detect
from deep_translator import GoogleTranslator

def translate_to_english(text):
    try:
        lang = detect(text)

        if lang == "en":
            return text

        translated = GoogleTranslator(source=lang, target="en").translate(text)
        return translated

    except:
        return text  # fallback if detection fails
```

```python
integrated_df.title = integrated_df.title.apply(translate_to_english)
```

# Standarized the JOb type value

```python
def clean_employment_type(text):
    text = text.lower()

    # Replace separators with spaces
    text = text.replace('-', ' ')
    return text

integrated_df.job_type = integrated_df.job_type.apply(clean_employment_type)

```

# Reconstruct a complete job description following the required format

```python

def build_embedding_text(row):
    return f"""
            Job Title: {row['title']}
            Company: {row['company']}
            Department : {row["Department"]}

            
            Job Details:
            - Job Type: {row['job_type']}
            - Posted since : {row["posted_since"]}
            - Domain: {row['field_of_work']}
            - Location: {row['location']}
            
            Responsibilities & Requirements:
            {row['clean_description']}
        """.strip()
    
integrated_df["description_improved"] = integrated_df[
    [
        'title', 'field_of_work', 'company',
        'job_type','Department',"posted_since",
        'location', 'clean_description'
    ]
].apply(build_embedding_text, axis=1)
```
